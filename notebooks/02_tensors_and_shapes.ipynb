# Lesson 2: Tensors and Shapes

Everything in a transformer is a tensor. In this notebook you create tensors, read and
change their shapes, and build the habit of printing the shape after every step.

Run each cell with Shift+Enter and watch the shapes that print.

````python
import torch

print("PyTorch version:", torch.__version__)
````

## Creating tensors

A tensor is a multi-dimensional array of numbers. Here are the ways you will make them
most often.

````python
a = torch.tensor([1, 2, 3])   # from a Python list
z = torch.zeros(2, 3)         # a 2 by 3 grid of zeros
o = torch.ones(2, 3)          # a 2 by 3 grid of ones
r = torch.randn(2, 3)         # random values from a normal distribution
n = torch.arange(6)           # 0, 1, 2, 3, 4, 5

print("a:", a)
print("z:", z)
print("r:", r)
print("n:", n)
````

## The shape is the whole game

Read `.shape` left to right. `torch.Size([2, 3])` means 2 rows, each holding 3 numbers.

````python
print("a.shape:", a.shape, "ndim:", a.ndim)
print("z.shape:", z.shape, "ndim:", z.ndim)
print("n.shape:", n.shape, "ndim:", n.ndim)
````

## The (batch, seq, dim) convention

Almost every tensor in this course has this shape:

```
(batch, seq, dim)
```

- batch: how many sequences at once
- seq: how many tokens per sequence
- dim: how many numbers per token

A batch of 2 sentences, 4 tokens each, 8 numbers per token, is shape (2, 4, 8).

````python
x = torch.randn(2, 4, 8)   # (batch, seq, dim)
print("x.shape:", x.shape)
print("batch =", x.shape[0], "| seq =", x.shape[1], "| dim =", x.shape[2])
````

## Indexing and slicing

Each index picks a position along one axis. A colon `:` means "all of this axis".

````python
print("x[0].shape      ->", x[0].shape)        # first sequence
print("x[0, 1].shape   ->", x[0, 1].shape)     # second token of first sequence
print("x[0, 1, 2]      ->", x[0, 1, 2].item()) # a single number
print("x[:, 0, :].shape ->", x[:, 0, :].shape) # first token of every sequence
````

## Reshaping

Same numbers, different shape. `view` and `reshape` rearrange; `transpose` swaps two
axes; `unsqueeze` adds an axis of length 1.

````python
n = torch.arange(12)
print("n.shape        ->", n.shape)
print("view(3,4).shape->", n.view(3, 4).shape)

print("x.shape            ->", x.shape)
print("transpose(1,2)     ->", x.transpose(1, 2).shape)

v = torch.randn(8)
print("v.shape            ->", v.shape)
print("v.unsqueeze(0)     ->", v.unsqueeze(0).shape)
````

## Broadcasting

PyTorch stretches a smaller tensor to match a larger one when their shapes are
compatible. Rule, read from the right: axes match if they are equal, or one is 1.

````python
m = torch.ones(3, 4)
bias = torch.tensor([1.0, 2.0, 3.0, 4.0])   # shape (4,)
print("m + bias =")
print(m + bias)   # bias added to each of the 3 rows
````

## Reductions

Collapse an axis with sum or mean to pool information. `dim=-1` is the last axis;
`keepdim=True` keeps the collapsed axis as length 1.

````python
print("x.sum()                    ->", x.sum().item())
print("x.mean(dim=-1).shape        ->", x.mean(dim=-1).shape)
print("x.mean(dim=-1, keepdim).shape ->", x.mean(dim=-1, keepdim=True).shape)
````
