```python
import torch
import torch.nn.functional as F
```

## Activations are elementwise

Each one runs on every number on its own.

```python
x = torch.tensor([-2.0, -1.0, 0.0, 1.0, 2.0])
print("relu:   ", F.relu(x))
print("sigmoid:", torch.sigmoid(x))
print("tanh:   ", torch.tanh(x))
print("gelu:   ", F.gelu(x))
```

## Why we need them: two linear steps collapse into one

```python
A = torch.randn(4, 3)
B = torch.randn(2, 4)
x = torch.randn(3)
print("no activation, still linear:", torch.allclose(B @ (A @ x), (B @ A) @ x))
```

## A ReLU is not linear, which is why it breaks the collapse

A linear map obeys f(a + b) = f(a) + f(b). ReLU does not, so it cannot be folded into a single matrix.

```python
a = torch.tensor([2.0, -1.0])
b = torch.tensor([-3.0, 4.0])
print("relu(a) + relu(b) =", F.relu(a) + F.relu(b))   # tensor([2., 4.])
print("relu(a + b)       =", F.relu(a + b))            # tensor([0., 3.]) - different
print("so ReLU is NOT linear:", not torch.allclose(F.relu(a) + F.relu(b), F.relu(a + b)))
```

## Shapes stay the same

```python
x3 = torch.randn(2, 5, 8)   # (batch, seq, dim)
print(F.relu(x3).shape)     # unchanged
```
