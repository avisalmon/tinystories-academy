## A linear layer: weights and bias

`nn.Linear(in, out)` holds a weight of shape (out, in) and a bias of shape (out,).

```python
layer = nn.Linear(in_features=4, out_features=3)
print("weight:", layer.weight.shape)   # (3, 4)
print("bias:  ", layer.bias.shape)     # (3,)
```

## It transforms the last axis

```python
x = torch.randn(2, 4)        # (batch, in)
print(x.shape, "->", layer(x).shape)          # (2, 4) -> (2, 3)
x3 = torch.randn(2, 5, 4)    # (batch, seq, in)
print(x3.shape, "->", layer(x3).shape)        # (2, 5, 4) -> (2, 5, 3)
```

## Under the hood it is x @ W.T + b

```python
x = torch.randn(2, 4)
manual = x @ layer.weight.T + layer.bias
print("matches nn.Linear:", torch.allclose(manual, layer(x)))
```
