```python
import torch
import torch.nn as nn
```

## An embedding is a lookup table

`nn.Embedding(vocab_size, embed_dim)` holds a matrix of shape (vocab_size, embed_dim). Row i is the vector for token id i.

```python
embedding = nn.Embedding(num_embeddings=10, embedding_dim=4)
print("weight shape:", embedding.weight.shape)   # (10, 4)
print("row for token id 0:", embedding.weight[0])
```

## Look up tokens

Feed it token ids; it returns their rows.

```python
ids = torch.tensor([1, 5, 2])
vectors = embedding(ids)
print(ids.shape, "->", vectors.shape)   # (3,) -> (3, 4)
```

## A batch of sequences becomes (batch, seq, dim)

```python
ids = torch.tensor([[1, 5, 2, 9],
                    [3, 3, 7, 0]])   # (batch=2, seq=4)
vectors = embedding(ids)
print(ids.shape, "->", vectors.shape)   # (2, 4) -> (2, 4, 4)
```

## Same id, same vector

It is a lookup, so the same id always returns the same row.

```python
a = embedding(torch.tensor([3]))
b = embedding(torch.tensor([3]))
print("same vector:", torch.allclose(a, b))
```
