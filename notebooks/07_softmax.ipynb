# Lesson 7: Softmax

In Lesson 6 your network produced logits: raw, unbounded scores, one per class. Softmax turns those scores into a probability distribution: every value positive, all of them summing to 1. You will meet softmax three more times in this course (the training loss, attention, and word sampling), so it is worth understanding well right now.

## Setup

```python
import torch
import torch.nn.functional as F

torch.manual_seed(0)
```

## Softmax by hand, and with PyTorch

The recipe is two steps: exponentiate every score, then divide by the total. That is exactly what `F.softmax` does.

```python
z = torch.tensor([2.0, 1.0, 0.1])

# by hand
ex = torch.exp(z)
by_hand = ex / ex.sum()
print('by hand: ', by_hand)

# with PyTorch
p = F.softmax(z, dim=-1)
print('F.softmax:', p)
print('match:', torch.allclose(by_hand, p))
```

## It is a distribution: it sums to 1

Every entry is positive and they add up to 1, so the outputs read as probabilities. The exponential amplifies differences, so the largest score gets the largest share.

```python
print('probabilities:', p)
print('sum:', p.sum())          # tensor(1.)
print('argmax class:', p.argmax().item())
```

## Temperature: sharp or soft

Divide the scores by a temperature `T` before softmax. `T` below 1 sharpens the distribution (more confident), `T` above 1 flattens it (more uniform). This exact knob returns in scaled attention and in text generation.

```python
print('T = 0.5 (sharper):', F.softmax(z / 0.5, dim=-1))
print('T = 1.0 (plain):  ', F.softmax(z / 1.0, dim=-1))
print('T = 2.0 (flatter):', F.softmax(z / 2.0, dim=-1))
```

## Shift-invariance and numerical safety

Adding the same constant to every score does not change softmax. Libraries use this to subtract the max score before exponentiating, so `exp` never overflows. PyTorch does this for you.

```python
print(torch.allclose(F.softmax(z, dim=-1), F.softmax(z + 100, dim=-1)))   # True
```
